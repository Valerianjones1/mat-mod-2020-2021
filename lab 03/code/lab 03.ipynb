```python
import math
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt
x_0 = 39800
y_0 = 21400
a_1 = 0.42
b_1 = 0.68
c_1 = 0.59
h_1 = 0.43
a_2 = 0.301
b_2 = 0.7
c_2 = 0.502
h_2 = 0.4
t0 = 0
tmax = 1
dt = 0.05
t = np.arange(t0, tmax, dt)
def P_1(t):
    p_1 = np.sin(5*t + 1)
    return p_1
def Q_1(t):
    q_1 = np.cos(5*t + 2)
    return q_1
def S_1(f, t):
    s_1 = -a_1*f[0] - b_1*f[1] + P_1(t)
    s_2 = -c_1*f[0] - h_1*f[1] + Q_1(t)
    return s_1, s_2
v = np.array([x_0, y_0])
f_1 = odeint(S_1, v, t)
plt.plot(t, f_1)
plt.ylabel('Количество людей в армии')
plt.xlabel('Время')
plt.legend(['Армия X', 'Армия Y'])
```
